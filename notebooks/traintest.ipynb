```python
import os
from pathlib import Path
import shutil
import os
```

```python

path = 'C:\\Users\\Prakhar Jadaun\\Documents\\Celebal Tech\CLE\\Project\\Main\\labelme\\Labels'
path = path + '\\'
path = path.replace('\\','/')
# print(path)
# print(os.listdir(path))
def getNames():
    i = 1
    lst = []
    for filename in os.listdir(path):
        lst.append(filename)

        # filename= filename.split('.')
        # print(filename[0])
    return lst
```

```python
lst = getNames()
```

```python
# 
from sklearn.model_selection import train_test_split
```

```python
X,y = train_test_split(lst,test_size=0.2,random_state=42)
```

```python
# moving files to train and test directory
srclabel = 'labelme\Labels'
srcimg = 'labelme\Images'
trgtrain = 'train'
trgtest = 'val'

# copying x files to train directory
for fname in X:
    # copying the files to the 
    # destination directory
    imgname = fname.split('.')[0] + '.png'
    shutil.copy2(os.path.join(srcimg,imgname),trgtrain)
    shutil.copy2(os.path.join(srclabel,fname),trgtrain)
```

```python
for fname in y:
    imgname = fname.split('.')[0] + '.png'
    shutil.copy2(os.path.join(srcimg,imgname),trgtest)
    shutil.copy2(os.path.join(srclabel,fname),trgtest)
```
